```python
# trainsetÏùÑ trainset Í≥º valsetÏúºÎ°ú Î∂ÑÌï†
import os
import shutil
from sklearn.model_selection import train_test_split

# ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ÏÖã Ìè¥Îçî Í≤ΩÎ°ú
dataset_folder = './raw_train'

# Ïù¥ÎèôÌï† Ìè¥Îçî Í≤ΩÎ°ú
train_folder = './train'
val_folder =  './valid'

# ÌÅ¥ÎûòÏä§ Ìè¥Îçî Î™©Î°ù ÏÉùÏÑ±
class_folders = [folder for folder in os.listdir(dataset_folder) if os.path.isdir(os.path.join(dataset_folder, folder))]

# Í∞Å ÌÅ¥ÎûòÏä§ Ìè¥ÎçîÏùò Îç∞Ïù¥ÌÑ∞ Î≥µÏÇ¨ Î∞è Î∂ÑÌï†
for class_folder in class_folders:
    class_folder_path = os.path.join(dataset_folder, class_folder)

    # ÌÅ¥ÎûòÏä§ Ìè¥Îçî ÏïàÏùò ÌååÏùº Î™©Î°ù Í∞ÄÏ†∏Ïò§Í∏∞
    file_list = os.listdir(class_folder_path)

    # ÌõàÎ†® Î∞è Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞Î°ú ÎÇòÎàÑÍ∏∞
    train_files, val_files = train_test_split(file_list, test_size=0.2, random_state=42)

    # ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ Î≥µÏÇ¨
    for file_name in train_files:
        src = os.path.join(class_folder_path, file_name)
        dst = os.path.join(train_folder, class_folder, file_name)
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        shutil.copy(src, dst)

    # Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞ Î≥µÏÇ¨
    for file_name in val_files:
        src = os.path.join(class_folder_path, file_name)
        dst = os.path.join(val_folder, class_folder, file_name)
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        shutil.copy(src, dst)
```

```python
import os
import csv

# ÌÅ¥ÎûòÏä§ Î¶¨Ïä§Ìä∏
classes = ['dog', 'elephant', 'giraffe', 'guitar', 'horse', 'house', 'person']

# txt ÌååÏùºÏù¥ ÏûàÎäî Ìè¥Îçî Í≤ΩÎ°ú
txt_folder_path = '/content/runs/classify/predict2/labels'

# Î™®Îì† txt ÌååÏùº Í≤ΩÎ°úÎ•º Ï†ÄÏû•Ìï† Î¶¨Ïä§Ìä∏
all_txt_files = []

# Ìè¥Îçî ÎÇ¥Ïùò Î™®Îì† txt ÌååÏùºÏóê ÎåÄÌï¥ ÏûëÏóÖ ÏàòÌñâ
for filename in os.listdir(txt_folder_path):
    if filename.endswith(".txt"):
        txt_file_path = os.path.join(txt_folder_path, filename)
        all_txt_files.append(txt_file_path)

# order Í∏∞Ï§ÄÏúºÎ°ú ÌååÏùº Í≤ΩÎ°ú Ï†ïÎ†¨
all_txt_files.sort(key=lambda x: int(os.path.splitext(os.path.basename(x))[0]))

# CSV ÌååÏùºÎ°ú Ï∂úÎ†•
output_filename = "submission.csv"
with open(output_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['order', 'label'])  # Ìó§Îçî Ïì∞Í∏∞

    # Î™®Îì† txt ÌååÏùºÏóê ÎåÄÌï¥ ÏûëÏóÖ ÏàòÌñâ
    for txt_file_path in all_txt_files:
        # Í∞ÄÏû• ÎÜíÏùÄ Ï†ïÌôïÎèÑÎ•º Í∞ÄÏßÑ ÌÅ¥ÎûòÏä§ Ï∂îÏ∂ú
        with open(txt_file_path, 'r') as file:
            lines = file.readlines()
            max_accuracy = 0
            max_class = ''
            for line in lines:
                parts = line.split()
                accuracy = float(parts[0])
                class_name = parts[1]
                if accuracy > max_accuracy:
                    max_accuracy = accuracy
                    max_class = class_name

        # ÌÅ¥ÎûòÏä§Î•º Ï†ïÏàò ÎùºÎ≤®ÎßÅÏúºÎ°ú Î≥ÄÌôò
        label_index = classes.index(max_class)

        # Îç∞Ïù¥ÌÑ∞ Ïì∞Í∏∞
        order = int(os.path.splitext(os.path.basename(txt_file_path))[0])
        writer.writerow([order, label_index])
```
